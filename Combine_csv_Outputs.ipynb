```python
import pandas as pd
import numpy as np
```

# World Cup Files

```python
# World Cup Emotions
WC_e = pd.read_csv("Bert_emotes_WC.csv")
WC_e["Match"] = "FIFA World Cup 2022"
WC_e = WC_e[['Match', 'text', 'Prediction_1', 'Prediction_2']]
WC_e.head()
```

```python
# World Cup Hate Speech
WC_hs = pd.read_csv("Bert_hs2_WC.csv")
WC_hs["Match"] = "FIFA World Cup 2022"
WC_hs = WC_hs[['Match', 'text', 'Prediction_1', 'Prediction_2']]
WC_hs.head()
```

# Euro Files

```python
# Euros Emotions
Euros_e = pd.read_csv("Bert_emotions_Euros.csv")
Euros_e["Match"] = "UEFA European Championship 2024"
Euros_e = Euros_e[['Match', 'text', 'Prediction_1', 'Prediction_2']]
Euros_e.head()
```

```python
# Euros Hate Speech
Euros_hs = pd.read_csv("Bert_hs_Euros.csv")
Euros_hs["Match"] = "UEFA European Championship 2024"
Euros_hs = Euros_hs[['Match', 'text', 'Prediction_1', 'Prediction_2']]
Euros_hs.head()
```

# City Vs Tottenham Files

```python
# City Vs Tottenham Emotions
PL_e = pd.read_csv("Bert_emotions_City_Tottenham.csv")
PL_e["Match"] = "Premier League 2024"
PL_e = PL_e[['Match', 'text', 'Prediction_1', 'Prediction_2']]
PL_e.head()
```

```python
# City Vs Tottenham Hate Speech
PL_hs = pd.read_csv("Bert_hs_City_Tottenham.csv")
PL_hs["Match"] = "Premier League 2024"
PL_hs = PL_hs[['Match', 'text', 'Prediction_1', 'Prediction_2']]
PL_hs.head()
```

# Liverpool Vs Real Madrid Files

```python
# Liverpool Vs Real Madrid Emotions
CL_e = pd.read_csv("Bert_emotes_CL_final.csv")
CL_e["Match"] = "Champions League 2024"
CL_e = CL_e[['Match', 'text', 'Prediction_1', 'Prediction_2']]
CL_e.head()
```

```python
# Liverpool Vs Real Madrid Hate Speech
CL_hs = pd.read_csv("Bert_hs_CL_final.csv")
CL_hs["Match"] = "Champions League 2024"
CL_hs = CL_hs[['Match', 'text', 'Prediction_1', 'Prediction_2']]
CL_hs.head()
```

```python
# Merge BERT Emotion files
merged = pd.concat([WC_e, Euros_e, PL_e, CL_e])
merged.to_csv("Merged_Emotions.csv", index=False)
```

```python
# Merge BERT Hate Speech files
merged_hs = pd.concat([WC_hs, Euros_hs, PL_hs, CL_hs])
merged_hs.to_csv("Merged_Hate_Speech.csv", index=False)
```
